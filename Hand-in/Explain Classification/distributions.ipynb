```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# My approach to making the cluster-cohort histograms. I did it for each feature in the cluster. 

intervals = ['[0.00, 0.05)', '[0.05, 0.10)', '[0.10, 0.15)', '[0.15, 0.20)', '[0.20, 0.25)', '[0.25, 0.3)', '[0.3, 0.35)', '[0.35, 0.4)', '[0.4, 0.45)', '[0.45, 0.5)', 
             '[0.5, 0.55)', '[0.55, 0.6)', '[0.6, 0.65)', '[0.65, 0.7)', '[0.7, 0.75)', 
             '[0.75, 0.8)', '[0.8, 0.85)', '[0.85, 0.90)', '[0.90, 0.95)', '[0.95, 1.00)']

counts = [0,0,0,0,0,1, 10, 21, 28, 37, 32, 16, 12, 5, 3, 1, 1, 0, 0, 0]   #The distribution for one feature in a cluster

# Split the interval to use in histogram
lower_bounds = [float(interval.split(',')[0][1:]) for interval in intervals]
upper_bounds = [float(interval.split(',')[1][:-1]) for interval in intervals]
midpoints = [(lower + upper) / 2 for lower, upper in zip(lower_bounds, upper_bounds)]

# Cluster histogram
plt.figure(figsize=(8, 6))
plt.bar(midpoints, height=counts, width=0.02, align='center')
plt.xlabel('Interval')
plt.ylabel('Count')
plt.title('C8.1 pH Values', fontsize=20)   # Update title based on cluster and feature 

# Error cohort area
plt.axvspan(0.35, 0.55, color='gray', alpha=0.3, label='Leaf 3 Decision Boundary')   # Update title based on leaf

plt.xticks(midpoints, intervals, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
